```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

data = pd.read_csv('data_value.csv')
corel_satat, P_val = pearsonr(data['Price'], data['Value'])

plt.figure(figsize = (6, 6))
plt.scatter(data['Price'], data['Value'], color = 'Red')

plt.show()
print(f'Корелляция: {corel_satat}; Р = {P_val}')
```
